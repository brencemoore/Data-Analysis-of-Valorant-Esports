```python
# Converts xx% to 0.xx
def convert_percent_to_decimal(x):
  new_x = float(x.strip('%')) / 100
  return new_x


# Splits player name to actual name and team name
def split_suffix(text):
    teams = ['DRX', 'FNC', 'SEN', 'LEV', 'EDG', 'TH', 'G2', 'TE']
    for x in teams:
        if text.endswith(x):
            return text[:-len(x)], x
    return text, None

# Cleans datasets to the same format
def clean_dataframe(df, winning_team):


  df.columns = ['Player', 'Team', 'Rating', 'ACS', 'Kills', 'Deaths', 'Assits', 'K/D_Difference', 'KAST', 'ADR', 'Headshot_Percent', 'First_Kills', 'First_Deaths', 'FK/FD_Difference']
  df = df.drop(5)

  df[['Player', 'Team']] = df['Player'].apply(lambda x: pd.Series(split_suffix(x)))

  df['Deaths'] = df['Deaths'].map(lambda x: int(x.split(' ')[1]))
  df['KAST'] = df['KAST'].map(convert_percent_to_decimal)
  df['Headshot_Percent'] = df['Headshot_Percent'].map(convert_percent_to_decimal)
  df = df.astype({
    'Rating': 'float',
    'ACS': 'int',
    'Kills': 'int',
    'Assits': 'int',
    'K/D_Difference': 'int',
    'ADR': 'int',
    'First_Kills': 'int',
    'First_Deaths': 'int',
    'FK/FD_Difference': 'int'
  })
  df['Map_Win'] = df['Team'].str.contains(winning_team)
  df = df.reset_index()
  df = df.drop('index', axis=1)

  return df
```
